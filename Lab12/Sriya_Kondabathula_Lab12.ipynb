### Problem 1 - Basic Features of Graphs
Here we will create a complete graph of 5 vertices and determine basic features about the graph, such as the number of edges it has, the degree, etc. Refer to the documentation here for help: https://networkx.org/documentation/stable/tutorial.html.



```python
import networkx as nx

# Enter your answer under each step.

# Create a complete graph of 5 vertices
G = nx.complete_graph(5)

# Plot the graph
nx.draw(G, with_labels=True)

# How many edges in this graph?
print("Number of edges in the graph:", G.number_of_edges())

# What is the adjacency matrix for this graph?
adj_matrix = nx.adjacency_matrix(G)
print("Adjacency matrix: \n", adj_matrix.todense())

# What is the degree for every vertex?
print("Degree for every vertex:", G.degree())

# What are the neighbors of the vertex 0 in this graph?
print("Neighbors of vertex 0:", list(G.neighbors(0)))
```

```python
# Calculate the shortest path between any two vertices (all the pairs in the graph).
print("Shortest path between all pairs of vertices \n")
list(nx.shortest_path_length(G))
```

```python
# Calculate the distance between each pair of vertices (0 and 1, 1 and 2, 2 and 3, etc.).
distances = dict(nx.shortest_path_length(G))

# Print the distance between each pair of vertices
for source in distances:
    for target in distances[source]:
        if source < target:
            distance = distances[source][target]
            print(f"Distance between {source} and {target}: {distance}")
```

```python
# Calculate the diameter of the graph.
diameter = -1
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > diameter:
      diameter = distance
print(f"The diameter of the graph is: {diameter}")
```

```python
# Is the graph connected? Check to see if it is. 
nx.is_connected(G)
```

```python
# How many connected components are in the graph?
print(f"Connected components in the graph: {nx.number_connected_components(G)}")
```

###Problem 3 - Karate Club Graph

```python
# There are some existing graphs that you can load in the networkx.
# Let's play with karate_club graph. Create an instance of this graph.
G = nx.karate_club_graph()

# Draw it
nx.draw(G, with_labels=True, node_size = 500)
plt.show()
```

```python
# How many connected components are in this graph?
print(f"Connected components in the graph: {nx.number_connected_components(G)}")
```

```python
# Print the degree of each vertex in this graph
print("Degree of each vertex:", dict(G.degree()))
```

```python
# What is the diameter of this graph?
diameter = -1
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > diameter:
      diameter = distance
print(f"The diameter of the graph is: {diameter}")
```

```python
# What is the path with the distance equal to this diameter?
# Hint: Longest distance, nx.shortest_path(G)

# Find the pair of nodes with the longest distance
max_distance = -1
max_pair = None

# Add more code here.... 
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance > max_distance:
      max_distance = distance
      max_pair = (source_node, target_node)
print(f"Pair of nodes with longest distance is : {max_pair}")

# Get the path corresponding to the longest distance
print("Path corresponding to the longest distance", nx.shortest_path(G, max_pair[0], max_pair[1]))
```

```python
# There might be more than one path correponding to this distance, can you list them?

# Find the pair of nodes with the longest distance
max_distance = 5
max_pair_list = []

# Add more code here... 
for source_node, distances in nx.shortest_path_length(G):
  for target_node, distance in distances.items():
    if distance == max_distance:
      max_pair_list.append((source_node, target_node))
print(f"maximum distance paths: {len(max_pair_list)} and they are {max_pair_list}")
# Get the path corresponding to the longest distance
for source, target in max_pair_list:
  print(nx.shortest_path(G, source, target))
```
